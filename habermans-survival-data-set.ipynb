```python
import pandas as pd
import numpy as np
from pandas import read_csv
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as pt

names = ['age','op_year','axil_nodes','surv_status']

df = read_csv('haberman.csv',names = names)
```

```python
#Checking for the length of the dataframe
len(df)

# creating bins for age and op_year

df['age_bins'] = pd.cut(x=df['age'],bins=[25,30,35,40,45,50,55,60,65,70,75,80,85])

df['op_year_bins'] = pd.cut(x=df['op_year'],bins=[55,60,65,70])
```

```python
# Univariate :
#Plotting the data using sns to see how well each attribute is distributed . Using displot and violinplot . Its seen that
# for axil_nodes data is distributed towards the left side. Data is largely consentrated between 0-3 and then subsequently lesser
#post 5.

pt.subplot(4,4,4)

sns.distplot(df['age'],kde = False)
pt.title('Box plot for age')
pt.show()

sns.distplot(df['op_year'],kde = False)
pt.title("Box plot for op_year")
pt.show

pt.figure(2)
sns.distplot(df['axil_nodes'],kde = False)
pt.title("Box plot for axil_nodes")
pt.show

pt.figure(3)
sns.violinplot(df['axil_nodes'],kde = False)
pt.title("violinplot plot for axil_nodes")
pt.show()

```

```python
# Bivariate plotting :
#
pt.figure(figsize=(45,50),dpi=150,facecolor='w',edgecolor='b')
pt.subplot(4,4,4)
#sns.boxplot(x='age',y='op_year',hue='surv_status',data=df,)
#pt.show()

#sns.boxplot(x='age_bins',y='surv_status',data=df)
#pt.show()

# Survival rates have been better for age group 35-40 ,50-55 years i.e beyond 5 years between year 1955-60. Between
#1960-1965 recovery rates have been good for age groups between 40-65 and 1970-75 the survival rates are good for
# age group 45-65 years.
sns.boxplot(y='op_year_bins',x='surv_status',data=df,hue='age_bins')

df.groupby(['surv_status','op_year_bins','age_bins']).age_bins.count()
```

```python
# Axil nodes are highest for age group between 30-35 years , followed by 50-55 and 40-45 years age group.
sns.barplot(x='axil_nodes',y='age_bins',data=df)
```
